```python
import os
import glob

# مسیر هدف
path = r"C:\Users\Saber\Documents\GitHub\rezaeisaber.github.io\assets\img\vol"

# بررسی وجود مسیر
if os.path.exists(path):
    # جستجو برای تمام فایل‌های تصویری (مثلاً .jpg, .png, .jpeg) در مسیر داده شده و زیرپوشه‌ها
    image_files = glob.glob(os.path.join(path, "**", "*.jpg"), recursive=True) + \
                  glob.glob(os.path.join(path, "**", "*.png"), recursive=True) + \
                  glob.glob(os.path.join(path, "**", "*.jpeg"), recursive=True)

    # دیکشنری برای گروه‌بندی فایل‌ها بر اساس پوشه
    folder_dict = {}

    # دسته‌بندی فایل‌ها بر اساس پوشه‌ها
    for image in image_files:
        folder_name = os.path.basename(os.path.dirname(image))  # نام پوشه
        image_name = os.path.basename(image)  # نام فایل
        if folder_name not in folder_dict:
            folder_dict[folder_name] = []
        folder_dict[folder_name].append(image_name)

    # چاپ آدرس‌ها به فرمت مورد نظر
    for folder, images in folder_dict.items():
        # چاپ نام فولدر
        print(f"Folder: {folder}")
        
        # چاپ آدرس تصاویر داخل فولدر به فرمت ../assets/img/vol/[folder_name]/[picture_name].jpg
        for image in images:
            print(f"../assets/img/vol/{folder}/{image}")
        
        # چاپ خط‌کش بین فولدرها
        print("--------")
else:
    print(f"مسیر '{path}' پیدا نشد.")
```
